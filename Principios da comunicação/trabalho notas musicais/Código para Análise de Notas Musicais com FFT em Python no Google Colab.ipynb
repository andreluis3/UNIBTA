```python
#Nota 1
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.signal import find_peaks

# Caminho real do áudio da Nota 1
caminho_audio = '/content/drive/MyDrive/Colab Notebooks/Notas musicais/Nota 1.wav'

# Leitura do áudio
taxa_amostragem, dados = wavfile.read(caminho_audio)

# Se estéreo, usa só um canal
if len(dados.shape) == 2:
    dados = dados[:, 0]

# FFT
N = len(dados)
frequencias = np.fft.rfftfreq(N, 1 / taxa_amostragem)
espectro = np.abs(np.fft.rfft(dados))

# Encontrar picos significativos
picos, _ = find_peaks(espectro, height=np.max(espectro)*0.1)
harmonicas_freqs = frequencias[picos]
harmonicas_vals = espectro[picos]

# Ordenar picos por magnitude
indices_ordenados = np.argsort(harmonicas_vals)[::-1]
freqs_principais = harmonicas_freqs[indices_ordenados]

# Fundamental
fundamental = freqs_principais[0]

# Harmônicas desejadas: 2ª, 3ª, 5ª e 6ª
harmonicas_desejadas = [2, 3, 5, 6]
harmonicas_freqs_aprox = [fundamental * i for i in harmonicas_desejadas]

harmonicas_real = []
harmonicas_mags = []
for freq in harmonicas_freqs_aprox:
    idx = np.argmin(np.abs(frequencias - freq))
    harmonicas_real.append(frequencias[idx])
    harmonicas_mags.append(espectro[idx])

# Exibir resultados
print(f"Fundamental: {fundamental:.2f} Hz")
for i, h in enumerate(harmonicas_real):
    print(f"{harmonicas_desejadas[i]}ª Harmônica: {h:.2f} Hz")

# 🎵 GRÁFICO 1: Espectro completo
plt.figure(figsize=(14, 5))
plt.plot(frequencias, espectro, label='Frequência Fundamental e harmônicas nota 1')
plt.title('Frequência Fundamental e harmônicas nota 1')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.xlim(0, 5000)
plt.grid(True)
plt.tight_layout()
plt.show()

# 🎯 GRÁFICO 2: Fundamental e harmônicas
plt.figure(figsize=(10, 5))
plt.stem([fundamental] + harmonicas_real,
         [espectro[np.argmin(np.abs(frequencias - fundamental))]] + harmonicas_mags,
         linefmt='g-', markerfmt='ro', basefmt='k-')
plt.title('Espectro de frequência - Nota 1')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.xticks([fundamental] + harmonicas_real,
           ['Fundamental'] + [f'{n}ª Harmônica' for n in harmonicas_desejadas])
plt.grid(True)
plt.tight_layout()
plt.show()
```
